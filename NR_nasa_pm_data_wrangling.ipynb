```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_train=pd.read_csv('PM_train.txt',sep=' ',header=None)
dataset_train.head()
```

```python
dataset_train=dataset_train.drop([26,27],axis=1)
```

```python
col_names = ['id','cycle','setting1','setting2','setting3','s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
```

```python
dataset_train.columns=col_names
dataset_train.head()
```

```python
dataset_test=pd.read_csv('PM_test.txt',sep=' ',header=None).drop([26,27],axis=1)
dataset_test.columns=col_names
dataset_test.head()
```

```python
pm_truth=pd.read_csv('PM_truth.txt',sep=' ',header=None).drop([1],axis=1)
pm_truth.columns=['more']
pm_truth['id']=pm_truth.index+1
pm_truth.head()
```

```python
pm_truth.shape
```

```python
# generate column max for test data
rul = pd.DataFrame(dataset_test.groupby('id')['cycle'].max()).reset_index()
rul.columns = ['id', 'max']
rul.head()
```

```python
pm_truth['rtf']=pm_truth['more']+rul['max']
pm_truth.head()
```

```python
pm_truth.drop('more', axis=1, inplace=True)
pm_truth.head()
```

```python
dataset_test=dataset_test.merge(pm_truth,on=['id'],how='left')
dataset_test['ttf']=dataset_test['rtf'] - dataset_test['cycle']
dataset_test.drop('rtf', axis=1, inplace=True)
dataset_test.head()
```

```python
df_train=dataset_train.copy()
df_test=dataset_test.copy()
```

```python
df_train['ttf'] = df_train.groupby(['id'])['cycle'].transform(max)-df_train['cycle']
```

```python
period=30
df_train['label_bnc'] = df_train['ttf'].apply(lambda x: 1 if x <= period else 0)
df_train['label_mcc'] = df_train['ttf'].apply(lambda x: 2 if x <= period/2 else 1 if x <= period else 0)
df_train.head()
```

```python
df_test['label_bnc'] = df_test['ttf'].apply(lambda x: 1 if x <= period else 0)
df_test['label_mcc'] = df_test['ttf'].apply(lambda x: 2 if x <= period/2 else 1 if x <= period else 0)
df_test.head()
```

```python
df_test.label_mcc.sum()
```

```python
sensor=['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']
```

```python
def avg_std(df,r_window):
    for s in sensor:
        col1=s+'_av'
        col2=s+'_std'
        df[col1]=df.groupby('id')[s].rolling(window=r_window,min_periods=1).mean().values
        df[col2]=df.groupby('id')[s].rolling(window=r_window,min_periods=1).std().fillna(0).values
    return df
```

```python
df_train=avg_std(df_train,10)
df_train.head()
```

```python
df_test=avg_std(df_test,10)
df_test.head()
```
